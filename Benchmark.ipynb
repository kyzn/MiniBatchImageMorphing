```python
import mnist_loader
import network2
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np

def ANN(name):
    
    epochs=30
        
    #Load the data file
    trn, val, tst = mnist_loader.load_data_wrapper()

    #Initialize the network and call SGD
    net = network2.Network([784,90,10],cost=network2.CrossEntropyCost) 
    net.large_weight_initializer()

    tst_cst, tst_acc, trn_cst, trn_acc =\
    net.SGD(training_data=trn, evaluation_data=tst, epochs=epochs, 
            mini_batch_size=10, eta=0.5, lmbda=0.1,
            monitor_evaluation_cost=True,
            monitor_evaluation_accuracy=True,
            monitor_training_cost=True,
            monitor_training_accuracy=True)

    #Write numbers to a file 
    fo=open("results.txt","a")
    np.savetxt(fo,np.array((tst_cst, tst_acc, trn_cst, trn_acc)),delimiter=',',header=name)
    fo.close()

    #Draw plots
    plt.plot(xrange(0,epochs),tst_acc,label="Test Accuracy")
    plt.plot(xrange(0,epochs),trn_acc,label="Training Accuracy")
    plt.title("Accuracy: "+name)
    plt.legend(loc=0)
    plt.show()

    plt.plot(xrange(0,epochs),tst_cst,label="Test Cost")
    plt.plot(xrange(0,epochs),trn_cst,label="Training Cost")
    plt.title("Cost: "+name)
    plt.legend(loc=0)
    plt.show()
```

```python
ANN("Benchmark, no morphing")
```
